```python
# TUGAS 1: Eksplorasi Array dan Matrix
import numpy as np

# 1. Array 20 angka random 0-100
arr = np.random.randint(0, 101, 20)
print("Array:", arr)

# 2. Statistik dasar
print("\nMean:", np.mean(arr))
print("Median:", np.median(arr))
print("Std Dev:", np.std(arr))
print("Min:", np.min(arr))
print("Max:", np.max(arr))

# 3. Matrix 5x5 random
matrix = np.random.randint(0, 10, (5, 5))
print("\nMatrix 5x5:\n", matrix)

# 4. Transpose
print("\nTranspose:\n", matrix.T)

# 5. Determinant
print("\nDeterminant:", np.linalg.det(matrix))
```

```python
# TUGAS 2: Analisis Data dengan Pandas
import pandas as pd

# Data 10 mahasiswa
data = {
    "NIM": ["001", "002", "003", "004", "005", "006", "007", "008", "009", "010"],
    "Nama": ["Andi", "Budi", "Citra", "Dewi", "Eka", "Fajar", "Gina", "Hadi", "Intan", "Joko"],
    "Matematika": [85, 60, 90, 72, 55, 95, 68, 89, 91, 58],
    "Fisika": [80, 62, 92, 70, 50, 94, 65, 88, 90, 60],
    "Kimia": [82, 58, 89, 75, 53, 93, 70, 85, 87, 55]
}
df = pd.DataFrame(data)
print("Data Mahasiswa:")
print(df)

# 1. Rata-rata per mata kuliah
print("\nRata-rata per Mata Kuliah:")
print(df[["Matematika", "Fisika", "Kimia"]].mean())

# 2. Rata-rata per mahasiswa
df["Rata-rata"] = df[["Matematika", "Fisika", "Kimia"]].mean(axis=1)
print("\nRata-rata per Mahasiswa:")
print(df[["Nama", "Rata-rata"]])

# 3. 5 mahasiswa dengan nilai tertinggi
print("\nTop 5 Mahasiswa:")
print(df.sort_values("Rata-rata", ascending=False).head())

# 4. Filter mahasiswa dengan rata-rata >= 80
print("\nMahasiswa dengan rata-rata >= 80:")
print(df[df["Rata-rata"] >= 80])

# 5. Export ke CSV
df.to_csv("data_mahasiswa.csv", index=False)
print("\nData berhasil diekspor ke 'data_mahasiswa.csv'")
```

```python
# TUGAS 3: Visualisasi Data
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(14, 10))

# 1. Bar Chart - Rata-rata Nilai per Mata Kuliah
plt.subplot(2, 2, 1)
rata2 = df[["Matematika", "Fisika", "Kimia"]].mean()
bars = plt.bar(rata2.index, rata2.values, width=0.6, edgecolor='black')
plt.title("Rata-rata Nilai per Mata Kuliah", fontsize=14, fontweight='bold')
plt.ylabel("Nilai", fontsize=12)
plt.ylim(70, 80)  # Mulai dari 70 agar tidak dominan di atas
plt.yticks(np.arange(70, 81, 2))
plt.grid(axis="y", linestyle="--", alpha=0.3)

# 2. Histogram - Distribusi Nilai Matematika
plt.subplot(2, 2, 2)
plt.hist(df["Matematika"], bins=5, rwidth=0.7, edgecolor='black')
plt.title("Distribusi Nilai Matematika", fontsize=14, fontweight='bold')
plt.xlabel("Nilai", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.xticks([55, 60, 65, 70, 75, 80, 85, 90, 95])
plt.grid(axis="y", linestyle="--", alpha=0.3)

# 3. Scatter Plot - Matematika vs Fisika
plt.subplot(2, 2, 3)
plt.scatter(df["Matematika"], df["Fisika"], alpha=0.6)
plt.title("Hubungan Matematika vs Fisika", fontsize=14, fontweight='bold')
plt.xlabel("Matematika", fontsize=12)
plt.ylabel("Fisika", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.3)

# 4. Boxplot - Semua Nilai
plt.subplot(2, 2, 4)
df[["Matematika", "Fisika", "Kimia"]].plot(kind="box", ax=plt.gca(), legend=False)
plt.title("Boxplot Nilai Mata Kuliah", fontsize=14, fontweight='bold')
plt.grid(axis="y", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# TUGAS 4: Eksplorasi Dataset
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset wine
wine = load_wine(as_frame=True)
wine_df = wine.frame

# 1. Tampilkan 10 baris pertama
print("10 Baris Pertama Dataset Wine:")
print(wine_df.head(10))

# 2. Statistik deskriptif
print("\nStatistik Deskriptif:")
print(wine_df.describe())

# 3. Visualisasi 1: Histogram Alcohol
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(wine_df["alcohol"], bins=10, rwidth=0.7, edgecolor='black')
plt.title("Distribusi Alcohol", fontsize=14, fontweight='bold')
plt.xlabel("Alcohol", fontsize=12)
plt.ylabel("Frekuensi", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.3)

# Visualisasi 2: Scatter Alcohol vs Malic Acid
plt.subplot(1, 2, 2)
plt.scatter(wine_df["alcohol"], wine_df["malic_acid"], alpha=0.6)
plt.title("Alcohol vs Malic Acid", fontsize=14, fontweight='bold')
plt.xlabel("Alcohol", fontsize=12)
plt.ylabel("Malic Acid", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()

# 4. Interpretasi singkat
print("\n" + "="*50)
print("INTERPRETASI DATASET WINE")
print("="*50)
print("""
1. Dataset Wine memiliki 178 sampel dengan 13 fitur kimiawi dari tiga jenis wine.
2. Kadar alcohol berkisar antara 11-15%, dengan distribusi yang relatif normal.
3. Terdapat korelasi negatif antara alcohol dan malic acid, di mana wine dengan alcohol tinggi cenderung memiliki malic acid rendah.
4. Dataset ini cocok untuk klasifikasi jenis wine berdasarkan kandungan kimianya.
""")
```
